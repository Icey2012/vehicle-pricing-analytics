```python
import pandas as pd
from ydata_profiling import ProfileReport
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('../data/raw/original_data.csv',index_col=0)
```

```python
profile=ProfileReport(df,title='Vehicle_Price_Analytics_EDA_Report')
profile.to_file("my_report.html")
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
import numpy as np

# 用所有特征（包括分类变量，直接当成数值用）
X = df[['Horsepower', 'Engine_CC', 'Manufacture_Year']].values
y = df['Price_USD'].values

tree = DecisionTreeRegressor(max_depth=4)
scores = cross_val_score(tree, X, y, cv=5, scoring='r2')
print(f"R²: {np.mean(scores):.3f}")
```

```python
# 构造：马力 × 品牌（豪华品牌标记）
df['is_luxury'] = df['Brand'].isin(['Mercedes-Benz', 'BMW', 'Audi']).astype(int)
df['hp_x_luxury'] = df['Horsepower'] * df['is_luxury']

# 看这个交互特征与价格的相关性
print(df['hp_x_luxury'].corr(df['Price_USD']))
```
